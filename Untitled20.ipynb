```python

from tensorflow.keras.preprocessing.sequence import pad_sequences
max_seq_length = 180  # 最大序列长度
#使用 TensorFlow 中的 pad_sequences 函数将 data 列表中的窗口数据进行填充
#使它们的长度都等于 max_seq_length
#填充的方式采用 post
#填充值为0，填充后的数据存储在变量 padded_data 中
padded_data = pad_sequences(data, maxlen=max_seq_length, padding='post', value=0.0)

# 运用tf.data.Dataset创建一个数据集 real_series
#将填充后的数据作为输入数据
#shuffle 函数对数据集进行了随机打乱
#batch 函数将数据集分成了多个批次，每个批次的大小为 batch_size
real_series = (tf.data.Dataset.from_tensor_slices(padded_data)
               .shuffle(buffer_size=n_windows)
               .batch(batch_size))
#使用 iter 函数将数据集转换成了一个可迭代对象 real_series_iter，该对象可以用于训练模型
real_series_iter = iter(real_series.repeat())
#主要作用：将填充后的数据集转换为 TensorFlow 中的数据集对象，并对其进行了 shuffle 和 batch 处理
```

```python
#定义生成器函数 make_random_data，用于生成随机数据
def make_random_data():
    while True:#使用无限循环，表示每次调用该生成器函数时都会生成一批新的随机数据
        yield np.random.uniform(low=0, high=1, size=(seq_len, n_seq))
#随机数据的生成使用了 NumPy 库中的 np.random.uniform 函数
#low 和 high 参数分别表示随机数的最小值和最大值，size 参数表示数组的形状，即数据的长度和特征数
```

```python
#从生成器函数 make_random_data 中生成随机数据，并将其转换为 TensorFlow 数据集对象
random_series = iter(tf.data.Dataset
                     .from_generator(make_random_data, output_types=tf.float32)
                     .batch(batch_size)
                     .repeat())
#from_generator 函数可以将一个生成器函数转换为 TensorFlow 数据集对象
#output_types 参数指定了生成器函数生成的数据类型
#batch 函数将生成的随机数据划分成多个批次，每个批次的大小为 batch_size
#repeat 函数将数据集重复多次，以便在训练模型时反复使用
#用 iter 函数将数据集转换为一个可迭代对象 random_series
#这段代码的作用是生成随机数据集
```

```python
#定义了一个函数 make_rnn，用于创建一个简单的循环神经网络（RNN）模型
#input_shape，表示输入数据的形状
#n_layers，表示模型中的 GRU 层的数量
#hidden_units，表示每个 GRU 层中的隐藏单元数
#output_units，表示模型输出层的单元数
def make_rnn(input_shape, n_layers, hidden_units, output_units, name):
    return Sequential([Input(shape=input_shape)] +      
                      [GRU(units=hidden_units,
                           return_sequences=True,
                           name=f'GRUvy_{i + 1}') for i in range(n_layers)] +
                      [Dense(units=output_units,
                             activation='sigmoid',
                             name='OUT')], name=name)
#该函数使用 Keras 中的 Sequential 类创建一个顺序模型，并在模型中添加了一个输入层，一个或多个 GRU 层，以及一个输出层
#模型中每个 GRU 层都使用了 return_sequences=True 参数，表示输出的序列数据将传递给下一层
#输出层使用了 sigmoid 激活函数，用于将输出数据映射到 [0, 1] 的范围内。
```

```python
embedder = tcn_ae.build_encoder()
recovery = tcn_ae.build_decoder()
```

```python
#这段代码用于对输入数据进行压缩和重构
H = embedder(X)#编码器模型 embedder 对输入数据 X 进行编码操作
X_tilde = recovery(H)#使用解码器模型 recovery 对编码后的低维表示 H 进行解码操作
#将输入数据 X 作为自编码器模型的输入,将重构后的数据 X_tilde 作为自编码器模型的输出
#通过 Model 函数构建自编码器模型
autoencoder = Model(inputs=X,
                    outputs=X_tilde,
                    name='Autoencoder')
autoencoder.summary()                   
```

```python
#这段代码实现的是一个用于训练自编码器模型的函数，其目的是通过反向传播算法来最小化输入数据与重构数据之间的重构误差，从而学习到一个能够对输入数据进行压缩和解压缩的编码器和解码器模型。
@tf.function
def train_autoencoder_init(x):
    with tf.GradientTape() as tape:#创建一个记录梯度信息的上下文环境，用于计算可训练变量的梯度
        x_tilde = autoencoder(x)#调用自编码器模型 autoencoder 对输入数据 x 进行编码和解码操作，得到重构后的输出数据 x_tilde        
        embedding_loss_t0 = mse(x, x_tilde)#计算重构误差，使用MSE损失函数度量输入数据x与重构后的数据x_tilde之间的差异
        e_loss_0 = 10 * tf.sqrt(embedding_loss_t0)
#使用平方根误差RMSE作为编码器的初始化损失函数，并乘以一个常数因子 10。这个常数因子的作用是放大损失函数的值，以加快初始化过程        
    var_list = embedder.trainable_variables + recovery.trainable_variables#获取编码器和解码器模型中所有可训练变量的列表
    gradients = tape.gradient(e_loss_0, var_list)#计算编码器和解码器模型中所有可训练变量关于初始化损失函数 e_loss_0 的梯度（反向传播算法）
    autoencoder_optimizer.apply_gradients(zip(gradients, var_list))#使用优化器更新编码器和解码器模型中所有可训练变量的值，以最小化损失函数
    return tf.sqrt(embedding_loss_t0)
```

```python
######看训练函数#####
#目的是训练自编码器的初始化过程
#训练方法是梯度下降法，运用Adam优化器训练
train_steps = 100
for step in tqdm(range(train_steps)):
    X_ = next(real_series_iter)#获取数据
    step_e_loss_t0 = train_autoencoder_init(X_)#运用自定义train_autoencoder_init训练函数来训练
    with writer.as_default():#用writer对象将自编码器的损失进行了记录
        tf.summary.scalar('Loss Autoencoder Init', step_e_loss_t0, step=step)
#tf.summary.scalar函数记录了名为“Loss Autoencoder Init”的标量摘要，并记录了当前步骤的step_e_loss_t0值。
```

```python
#定义训练函数train_supervisor（训练监督器模型）
@tf.function
def train_supervisor(x):    
    with tf.GradientTape() as tape:
        h = embedder(x)#首先将输入张量 x 传递给嵌入器模型 embedder
        h_hat_supervised = supervisor(h)#将 h 传递给监督器模型 supervisor
        g_loss_s = mse(h[:, 1:, :], h_hat_supervised[:, :-1, :])#计算监督器模型的均方误差损失
    var_list = supervisor.trainable_variables
    gradients = tape.gradient(g_loss_s, var_list)#使用gloss计算梯度
    supervisor_optimizer.apply_gradients(zip(gradients, var_list))#使用梯度下降算法更新监督器模型的权重
    return g_loss_s
```

```python
for step in tqdm(range(train_steps)):
    X_ = next(real_series_iter)#获取数据
    step_g_loss_s = train_supervisor(X_)#调用 train_supervisor 函数来训练监督器模型
    with writer.as_default():
        tf.summary.scalar('Loss Generator Supervised Init', step_g_loss_s, step=step)
```

```python
#定义了一个计算生成器的矩损失函数的函数
def get_generator_moment_loss(y_true, y_pred):
    #tf.nn.moments 函数计算了 y_true 和 y_pred 的均值和方差
    y_true_mean, y_true_var = tf.nn.moments(x=y_true, axes=[0])
    y_pred_mean, y_pred_var = tf.nn.moments(x=y_pred, axes=[0])
    #计算均值之间的差异 g_loss_mean 和方差之间的差异 g_loss_var
    #具体的，mean，var运用相减加绝对值，var加1e-6防止方差为0，保证计算稳定性
    g_loss_mean = tf.reduce_mean(tf.abs(y_true_mean - y_pred_mean))
    g_loss_var = tf.reduce_mean(tf.abs(tf.sqrt(y_true_var + 1e-6) - tf.sqrt(y_pred_var + 1e-6)))
    #将它们相加作为最终的生成器矩损失函数
    return g_loss_mean + g_loss_var
#解释：矩损失函数在GAN中运用较多，作为生成器的损失函数。矩损失函数比较生成数据和目标数据在某些特征上的统计分布，例如均值和方差。
#与传统损失函数比，矩损失函数衡量统计数据之间差异，传统损失函数衡量点之间的差异。
```

```python
#定义了一个训练生成器的函数 train_generator，具体训练方法就是运用梯度下降的方法降低loss
tf.function
def train_generator(x, z):#接受两个参数x，z，分别表示真实数据和生成器的输入噪声
    #函数的主要作用是计算生成器的损失函数，并根据损失函数的梯度更新生成器的参数。
    with tf.GradientTape() as tape:#记录生成器的前向计算过程，并计算生成器的损失函数
        y_fake = adversarial_supervised(z)
        #损失函数有4个部分
        generator_loss_unsupervised = bce(y_true=tf.ones_like(y_fake),
                                          y_pred=y_fake)
        #无监督损失函数：
        #使用二元交叉熵损失函数 bce 计算生成器输出的假数据 y_fake 和目标标签 tf.ones_like(y_fake) 之间的损失
        #更具体的，y_true是作为真实标签，相同维度但是数值全为1，代表真实数据概率为1
        #y_pred表示生成数据的概率，在0-1之间（因为输出通过一个 sigmoid 函数进行激活的）
        y_fake_e = adversarial_emb(z)
        generator_loss_unsupervised_e = bce(y_true=tf.ones_like(y_fake_e),
                                            y_pred=y_fake_e)
        #嵌入器对抗损失函数：
        #使用二元交叉熵损失函数 bce 计算生成器嵌入器输出的假数据 y_fake_e 和目标标签 tf.ones_like(y_fake_e) 之间的损失
        #理由同上，运用sigmoid函数，都在0-1之间
        h = embedder(x)
        h_hat_supervised = supervisor(h)
        generator_loss_supervised = mse(h[:, 1:, :], h_hat_supervised[:, 1:, :])
        #监督损失函数：
        #使用均方误差损失函数 mse 计算监督器的输出 h_hat_supervised 和目标标签 h[:, 1:, :] 之间的损失
        #[:, 1:, :]表示去掉第一个元素，因为第一个元素通常是一个起始符，不包含太多信息。
        #h是embedder的输出向量
        x_hat = synthesizer(z)
        generator_moment_loss = get_generator_moment_loss(x, x_hat)
        #生成器矩损失函数：
        #调用 get_generator_moment_loss 函数计算输入数据 x 和生成数据 x_hat 之间的矩损失函数
        generator_loss = (generator_loss_unsupervised +
                          generator_loss_unsupervised_e +
                          100 * tf.sqrt(generator_loss_supervised) +
                           100 * generator_moment_loss)
        #将四个部分加权相加得到生成器损失函数
    var_list = tcn_generator.trainable_variables + supervisor.trainable_variables
    gradients = tape.gradient(generator_loss, var_list)
    #计算损失函数 generator_loss 对于生成器的可训练参数的梯度
    generator_optimizer.apply_gradients(zip(gradients, var_list))
    #使用优化器 generator_optimizer 对生成器的参数进行更新
    return generator_loss_unsupervised, generator_loss_supervised, generator_moment_loss
```

```python
#定义了一个 train_embedder 函数,运用梯度下降算法训练
@tf.function
def train_embedder(x):#输入是一个张量 x
    with tf.GradientTape() as tape:
        h = embedder(x)#函数首先利用嵌入器模型 embedder 将输入数据 x 转换成嵌入向量 h
        h_hat_supervised = supervisor(h)#使用监督器模型 supervisor 处理嵌入向量 h，生成重建向量 h_hat_supervised
        generator_loss_supervised = mse(h[:, 1:, :], h_hat_supervised[:, 1:, :])
        #计算生成器的监督损失函数 generator_loss_supervised
        x_tilde = autoencoder(x)#使用自编码器模型 autoencoder 处理输入数据 x，得到重建数据 x_tilde
        embedding_loss_t0 = mse(x, x_tilde)#计算嵌入损失函数 embedding_loss_t0（mse）
        e_loss = 10 * tf.sqrt(embedding_loss_t0) + 0.1 * generator_loss_supervised
        #计算e_loss
    var_list = embedder.trainable_variables + recovery.trainable_variables
    gradients = tape.gradient(e_loss, var_list)
    #计算嵌入损失函数 e_loss 对嵌入器模型 embedder 和自编码器模型 recovery 可训练变量的梯度
    embedding_optimizer.apply_gradients(zip(gradients, var_list))#使用优化器 embedding_optimizer 对它们进行更新
    return tf.sqrt(embedding_loss_t0)
```

```python
#定义了一个判别器的损失函数get_discriminator_loss。
#它有两个参数x和z，分别表示真实样本和生成样本的输入。
def get_discriminator_loss(x, z):
    y_real = discriminator_model(x)#对真实样本进行预测
    discriminator_loss_real = bce(y_true=tf.ones_like(y_real),
                                  y_pred=y_real)#计算真实样本的损失（bce）
    #真实标签被设置为1（y_true）

    y_fake = adversarial_supervised(z)#对生成样本进行预测
    discriminator_loss_fake = bce(y_true=tf.zeros_like(y_fake),
                                  y_pred=y_fake)#计算生成样本的损失
    #真实标签被设置为0，表示这是一个生成样本

    y_fake_e = adversarial_emb(z)#对生成样本的嵌入向量进行预测
    discriminator_loss_fake_e = bce(y_true=tf.zeros_like(y_fake_e),
                                    y_pred=y_fake_e)#计算生成样本嵌入向量的损失
    
    #真实标签被设置为0
    #返回总体损失
    return (discriminator_loss_real +
            discriminator_loss_fake +
            gamma * discriminator_loss_fake_e)#gamma是一个超参数，用于权衡生成样本损失和生成样本嵌入向量损失的重要性
```

```python
#定义训练判别器的函数train_discriminator（梯度下降法）
@tf.function
def train_discriminator(x, z):#接受两个参数x和z，分别表示真实样本数据和生成器生成的样本数据
    with tf.GradientTape() as tape:
        discriminator_loss = get_discriminator_loss(x, z)
    #调用get_discriminator_loss(x, z)函数计算判别器的损失
    var_list = discriminator.trainable_variables
    #使用tape.gradient()计算损失对判别器可训练变量的梯度
    gradients = tape.gradient(discriminator_loss, var_list)
    #将梯度应用到判别器的可训练变量上，以更新判别器的参数
    #w=w-learning rate*gradient
    discriminator_optimizer.apply_gradients(zip(gradients, var_list))
    return discriminator_loss
```

```python
#表示当前训练过程中的最小损失值，初始化为正无穷大，用于记录训练过程中的最优模型参数
min_loss = np.inf
#表示在验证集上连续多少个epoch没有提升时，就停止训练
patience = 4
#表示停止训练的阈值，当验证集上的损失值减小的程度小于这个阈值时，认为模型已经收敛，可以提前停止训练
brk_crit = 1e-4
#表示当前连续多少个epoch没有提升，初始化为0。
#当验证集上的损失值连续多次没有提升时，wait会逐渐增加，直到达到patience时，就停止训练
wait = 0
```

```python
#定义了一个TCN编码层的前向传播函数
@tf.function
def tcn_enc_layer_call_and_return_conditional_losses(self, inputs, training=False):
    x = self.initial_conv(inputs)#首先对输入张量做一个卷积操作
    res_x = x#作为原始输入向量
    for i in range(self.repeats):
        x = self.res_blocks[i](x, training=training)#进行特征提取和非线性变换
        x += res_x#进行残差连接（元素对应相加）
        res_x = x#作为下一个残差进入循环，
    outputs = self.final_bn(x, training=training)#对输出结果进行归一化，使均值接近0，方差接近1
    return outputs
```

```python
#交替训练生成器，判别器
tf.config.run_functions_eagerly(True)
input = (180,200,6)#设置输入形状
train_steps = 300#训练步数
#定义变量，用于记录损失
step_g_loss_u = step_g_loss_s = step_g_loss_v = step_g_loss_r = step_e_loss_t0 = step_d_loss = 0
for step in range(train_steps):
    for kk in range(2):   
        #随机噪声向量数据集和真实数据集中分别获取一个随机噪声向量Z和一个真实时间序列数据X
        Z_ = next(random_series)
        X_ = next(real_series_iter)
        while X_.shape != Z_.shape:
            X_ = next(real_series_iter)
        #转换为float32型
        X_ = tf.cast(X_, tf.float32)
        Z_ = tf.cast(Z_, tf.float32)
        #调用train_generator函数训练生成器网络
        #记录生成器的重建损失、相似性损失和变化损失
        step_g_loss_u, step_g_loss_s, step_g_loss_v = train_generator(X_, Z_)
        #调用train_embedder函数训练嵌入器网络
        step_e_loss_t0 = train_embedder(X_)
    #从真实数据集和随机噪声向量数据集中分别获取一个时间序列数据X和一个随机噪声向量Z
    X_ = next(real_series_iter)
    Z_ = next(random_series)
    #调用get_discriminator_loss函数计算鉴别器的损失函数，并记录到step_d_loss变量中
#如果鉴别器的损失函数大于0.15，则调用train_discriminator函数训练鉴别器网络，并更新鉴别器网络的权重
#这里损失函数大于0.15可以认为鉴别器过拟合，它认为自己已经学会如何判别真假数据，实际并没有
    step_d_loss = get_discriminator_loss(X_, Z_)
    if step_d_loss > 0.15:
        step_d_loss = train_discriminator(X_, Z_)
#每100个训练步骤，输出当前训练步骤的鉴别器损失、生成器的重建损失、相似性损失和变化损失以及嵌入器网络的损失函数
    if step % 100 == 0: 
        print(f'{step:6,.0f} | d_loss: {step_d_loss:6.4f} | g_loss_u: {step_g_loss_u:6.4f} | '
              f'g_loss_s: {step_g_loss_s:6.6f} | g_loss_v: {step_g_loss_v:6.4f} | e_loss_t0: {step_e_loss_t0:6.4f}')
        
        wait += 1
        if min_loss - step_g_loss_s >= brk_crit:
          min_loss =  step_g_loss_s
          wait = 0
        if wait >= patience:
          break
        
    with writer.as_default():
        tf.summary.scalar('G Loss S', step_g_loss_s, step=step)
        tf.summary.scalar('G Loss U', step_g_loss_u, step=step)
        tf.summary.scalar('G Loss V', step_g_loss_v, step=step)
        tf.summary.scalar('E Loss T0', step_e_loss_t0, step=step)
        tf.summary.scalar('D Loss', step_d_loss, step=step)                                 
synthesizer.save(log_dir / 'synthesizer')  
```
